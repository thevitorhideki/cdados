Se isto ocorrer, a probabilidade de que os pre√ßos dos transportes coletivos tamb√©m aumentem √© de 0,5; caso contr√°rio, esta probabilidade √© de 0,1. 

Se naquele m√™s o pre√ßo das passagens de fato subirem, qual a probabilidade de os pre√ßos dos combust√≠veis n√£o terem sofrido majora√ß√£o?

**DEFINA OS EVENTOS APRESENTADOS NO ENUNCIADO:**

 * $C^c$: n√£o aumento do pre√ßo do combust√≠vel
 * $T$: aumento do pre√ßo do transporte
 * $T^c$: n√£o aumento do pre√ßo do transporte

**DEFINA AS PROBABILIDADES APRESENTADAS NO ENUNCIADO EM TERMOS DOS EVENTOS:**

 * $P(C)=0.4$
 * $P(T|C)=0.5$
 * $P(T|C^c)=0.1$

```python
# Enunciado
P_C = 0.4
P_T_dado_C = 0.5
P_T_dado_Cc = 0.1
```

**QUAL A PROBABILIDADE SOLICITADA NO ENUNCIADO EM TERMOS DOS EVENTOS:**
 
 * $P(C^c|T)=?$


**Resolu√ß√£o:**

A probabilidade condicional solicitada √© dada por

$P(C^c|T) = \frac{P(T \cap C^c)}{P(T)}$

Logo, primeiro √© necess√°rio calcular $P(T \cap C^c)$ e $P(T)$.

**Deixe claro todo o racioc√≠nio usado para chegar na probabilidade questionada:**

### EXERC√çCIO 2:

Uma revendedora de ve√≠culos trabalha com duas marcas de autom√≥veis: A e B. 70% dos que adquirem carros populares, escolhem a marca A. Dentre os que adquirem carros n√£o populares, 80% compram A. Sabe-se que 60% das vendas s√£o de carros populares. 

a. Qual √© a probabilidade de um consumidor comprar um carro da marca A?

b. Sabendo que uma pessoa comprou um carro da marca A, qual √© a probabilidade de ter sido um carro popular?
___

**DEFINA OS EVENTOS APRESENTADOS NO ENUNCIADO:**

 * $A$ : Carros da marca A
 * $P$ : Carros populares
 * $P^c$: Carros n√£o populares

**DEFINA AS PROBABILIDADES APRESENTADAS NO ENUNCIADO EM TERMOS DOS EVENTOS:**

 * $P(A | P)$ = 0.7
 * $P(A | P^c)$ = 0.8
 * $P(P)$ = 0.6

```python
P_A_dado_P = 0.7
P_A_dado_Pc = 0.8
P_P = 0.6
```

**QUAL A PROBABILIDADE SOLICITADA NO ENUNCIADO EM TERMOS DOS EVENTOS:**
 
 * a) $P(A)$
 * b) $P(P | A)$ = $\frac{P(P \cap A)}{P(A)}$


**Deixe claro todo o racioc√≠nio usado para chegar na probabilidade questionada:**

```python
# Exerc√≠cio a)
# Se P(P) = 0.6, ent√£o:
P_Pc = 1 - P_P

# Para calcular a interse√ß√£o de A e P:
P_AeP = P_A_dado_P * P_P

# Para calcular a interse√ß√£o de A e Pc:
P_AePc = P_A_dado_Pc * P_Pc

# Somando essas duas interse√ß√µes temos a probabilidade do carro comprado ser da marca A
P_A = P_AeP + P_AePc
print(f"A probabilidade do consumidor comprar um carro da marca A √© de {P_A*100}%")
```

```python
# Exerc√≠cio b)
# Sendo que uma pessoa comprou um carro da marca A, qual √© a probabilidade de ter sido um carro popular?

# Para isso √© necess√°rio saber a interse√ß√£o de P e A e dividir pela probabilidade de ser um carro A
P_P_dado_A = P_AeP / P_A
print(f"A probabilidade de um carro comprado ser popular dado que √© da marca A √© de {P_P_dado_A*100:.2f}%")
```

### EXERC√çCIO 3 - Cap√≠tulo 2 de Montgomery e Runger (2016):

Em uma opera√ß√£o de enchimento autom√°tico, a probabilidade de um enchimento incorreto ser√° 0,001, quando o processo for operado em baixa velocidade. Quando o processo for operado em alta velocidade, a probabilidade de um enchimento incorreto ser√° de 0,01.

Suponha de 30% das opera√ß√µes de enchimento autom√°tico em alta velocidade, e o restante seja cheio em baixa velocidade.

a. Qual a probabilidade de um reservat√≥rio ser cheio incorretamente?

b. Se um reservat√≥rio cheio incorretamente for encontrado, qual √© a probabilidade de que ele tenha sido cheio durante uma opera√ß√£o em alta velocidade?
___

**DEFINA OS EVENTOS APRESENTADOS NO ENUNCIADO:**

 * $E^c$ = Cheio incorretamente
 * $V$ = Alta velocidade
 * $V^c$ = Baixa velocidade

**DEFINA AS PROBABILIDADES APRESENTADAS NO ENUNCIADO EM TERMOS DOS EVENTOS:**

 * $P(E^c|V^e)$ = 0.001
 * $P(E^c|V)$ = 0.01
 * $P(V)$ = 0.3

```python
P_Ec_dado_Vc = 0.001
P_Ec_dado_V = 0.01
P_V = 0.3
```

**QUAL A PROBABILIDADE SOLICITADA NO ENUNCIADO EM TERMOS DOS EVENTOS:**
 
 * a) $P(E^c)$
 * b) $P(V|E^c)$ = $\frac{P(V \cap E^c)}{P(E^c)}$


**Deixe claro todo o racioc√≠nio usado para chegar na probabilidade questionada:**

```python
# Exerc√≠cio a)
# Para achar a probabilidade de um reservat√≥rio ter sido enchido incorretamente, primeiramente √© necess√°rio achar a probabilidade de interse√ß√£o de dois eventos: a probabilidade de o reservat√≥rio ter sido cheio incorretamente com a probabilidade de ter usado a velocidade baixa; a probabilidade de o reservat√≥rio ter sido cheio incorretamente com a probabilidade de ter usado a velocidade alta

# Se P(V) = 0.3, ent√£o:
P_Vc = 1 - P_V

# Achando as interse√ß√µes:
P_EceVc = P_Ec_dado_Vc * P_Vc
P_EceV = P_Ec_dado_V * P_V

# Somando as duas interse√ß√µes para achar a chance de ter sido cheio incorretamente
P_Ec = P_EceV + P_EceVc

print(f"A chance de um reservat√≥rio ter sido cheio incorretamente √© de {P_Ec * 100 :.2f}%")
```

```python
# Exerc√≠cio b)
# Para encontrar a probabilidade de o reservat√≥rio ter sido cheio em alta velocidade, dado que ele foi cheio incorretamente:
P_V_dado_Ec = P_EceV / P_Ec

print(f"Se um reservat√≥rio cheio incorretamente for encontrado, a chance ter sido cheio em alta velocidade √© {P_V_dado_Ec * 100 :.2f}%")
```

### EXERC√çCIO 5 - Cap√≠tulo 2 de Montgomery e Runger (2016):


Um sistema de codifica√ß√£o-decodifica√ß√£o consiste em tr√™s elementos: codifica, transmite e decodifica.

Uma codifica√ß√£o falha ocorre em 0,5% das mensagens processadas; erro de transmiss√£o ocorrem em 1% das mensagens;  e erro de decodifica√ß√£o ocorre em 0,1% das mensagens. Considere os erros como independentes.

a. Qual a probabilidade de se ter uma mensagem completamente livre de defeito?

b. Qual a probabilidade de uma mensagem ter tanto um defeito de codifica√ß√£o como de decodifica√ß√£o?
___

**DEFINA OS EVENTOS APRESENTADOS NO ENUNCIADO:**

 * $P(C^c)$: Probabilidade de ocorrer um erro de codifica√ß√£o
 * $P(T^c)$: Probabilidade de ocorrer um erro de transmiss√£o
 * $P(D^c)$: Probabilidade de ocorrer um erro de decodifica√ß√£o

**DEFINA AS PROBABILIDADES APRESENTADAS NO ENUNCIADO EM TERMOS DOS EVENTOS:**

 * $P(C^c)$ = 0.005
 * $P(T^c)$ = 0.01
 * $P(D^c)$ = 0.001

```python
P_Cc = 0.005
P_Tc = 0.01
P_Dc = 0.001
```

**QUAL A PROBABILIDADE SOLICITADA NO ENUNCIADO EM TERMOS DOS EVENTOS:**
 
 * a) $P(C \cap T \cap D)$
 * b) $P(C^c \cap D^c)$


**Deixe claro todo o racioc√≠nio usado para chegar na probabilidade questionada:**

```python
# Exerc√≠cio a)

# Para chegar na probabilidade de n√£o ocorrer nenhum erro, √© preciso encontrar os eventos complementares:
P_C = 1 - P_Cc
P_T = 1 - P_Tc
P_D = 1 - P_Dc

# Como eles s√£o independentes, basta multiplicar para encontrar a probabilidade:
P_CeTeD = P_C * P_T * P_D
print(f"A probabilidade de se ter uma mensagem completamente livre de defeito √© de {P_CeTeD * 100:.2f}%")
```

```python
# Exerc√≠cio b)

# Para chegar na probabilidade de uma mensagem ter tanto defeito de codifica√ß√£o quanto decodifica√ß√£o basta usar a adi√ß√£o de dois eventos:
P_CceDc = P_Cc * P_Dc
print(f"A probabilidade de uma mensagem ter tanto um defeito de codifica√ß√£o como de decodifica√ß√£o √© de {P_CceDc*100:.4f}%")
```
