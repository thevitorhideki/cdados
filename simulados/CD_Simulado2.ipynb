## PERGUNTA 5:

```python
not_us = df.loc[df['company_location'] != 'US', :]
not_us
```

```python
not_us['salary_in_10kusd'] = not_us['salary_in_usd']/10000
not_us
```

```python
amp = np.arange(0, 26, 2)

plt.figure(figsize=(20, 6))
plt.subplot(121)
plt.hist(not_us['salary_in_10kusd'], bins=amp, density=True, edgecolor="black")
plt.xlabel("Salário em 10k usd")
plt.ylabel("Densidade")
plt.xticks(amp)
plt.show()
```
